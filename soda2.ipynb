```python
import argparse
import ast
import json
import re
from typing import List, Dict, Tuple, Optional

import pandas as pd


def is_helper_role(name: str) -> bool:
    """Heuristic: identify counselor/therapist-like roles (case-insensitive)."""
    if not isinstance(name, str):
        return False
    name_norm = name.strip().lower()
    patterns = [
        r"therap",         # therapist / therapy / psychotherapist
        r"counsel",        # counselor / counseling
        r"counsell",       # UK spelling
        r"clinician",
        r"psychiatr",
        r"psycholog",
        r"social\s*worker",
    ]
    return any(re.search(p, name_norm) for p in patterns)



def parse_speakers(s: str) -> List[str]:
    """Parse speakers from numpy array format like "['Shan' 'Counselor' 'Shan' 'Counselor']" """
    if s is None:
        return []
    s = s.strip()
    
    # Handle numpy array format like "['Shan' 'Counselor' 'Shan' 'Counselor']"
    if s.startswith("['") and s.endswith("']"):
        # Remove the outer brackets and quotes
        inner = s[2:-2]  # Remove "['" and "']"
        # Replace newlines with spaces and split by "' '" pattern
        inner = inner.replace('\n', ' ')
        parts = inner.split("' '")
        return [part.strip() for part in parts if part.strip()]
    
    # Fallback to general parsing
    try:
        val = ast.literal_eval(s)
        if isinstance(val, (list, tuple)):
            return [str(x) for x in val]
    except Exception:
        pass
    
    return [s] if s else []


def parse_dialogue(s: str) -> List[str]:
    """Parse dialogue from JSON string format like '["text1"\\n \\'text2\\'\\n "text3"]' """
    if s is None:
        return []
    s = s.strip()
    
    # Handle dialogue format like '["text1"\\n \\'text2\\'\\n "text3"]'
    if s.startswith('["') and s.endswith('"]'):
        # Remove outer quotes and brackets
        inner = s[2:-2]  # Remove '["' and '"]'
        # Split by actual newline character, not escaped
        parts = inner.split('\n')
        result = []
        for part in parts:
            part = part.strip()
            if not part:
                continue
            # Remove leading/trailing quotes and spaces
            if part.startswith('"') and part.endswith('"'):
                part = part[1:-1]
            elif part.startswith("'") and part.endswith("'"):
                part = part[1:-1]
            # Unescape quotes
            part = part.replace("\\'", "'").replace('\\"', '"')
            if part.strip():
                result.append(part.strip())
        return result
    
    # Fallback to general parsing
    try:
        val = ast.literal_eval(s)
        if isinstance(val, (list, tuple)):
            return [str(x) for x in val]
    except Exception:
        pass
    
    return [s] if s else []


def parse_listlike(s: str) -> List[str]:
    """Legacy function - use parse_speakers or parse_dialogue instead"""
    if s is None:
        return []
    s = s.strip()
    
    # Handle numpy array format like "['Shan' 'Counselor' 'Shan' 'Counselor']"
    if s.startswith("['") and s.endswith("']"):
        # Remove the outer brackets and quotes
        inner = s[2:-2]  # Remove "['" and "']"
        # Split by "' '" pattern
        parts = inner.split("' '")
        return [part.strip() for part in parts if part.strip()]
    
    # Handle dialogue format like '["text1"\n \'text2\'\n "text3"]'
    if s.startswith('["') and s.endswith('"]'):
        # Remove outer quotes and brackets
        inner = s[2:-2]  # Remove '["' and '"]'
        # Split by newline pattern and clean up
        parts = inner.split('\\n')
        result = []
        for part in parts:
            part = part.strip()
            if not part:
                continue
            # Remove leading/trailing quotes and spaces
            if part.startswith('"') and part.endswith('"'):
                part = part[1:-1]
            elif part.startswith("'") and part.endswith("'"):
                part = part[1:-1]
            # Unescape quotes
            part = part.replace("\\'", "'").replace('\\"', '"')
            if part.strip():
                result.append(part.strip())
        return result
    
    # Handle standard Python list format
    try:
        val = ast.literal_eval(s)
        if isinstance(val, (list, tuple)):
            return [str(x) for x in val]
    except Exception:
        pass
    
    # Handle JSON-like format
    if s.startswith('[') and s.endswith(']'):
        inner = s[1:-1].strip()
        if not inner:
            return []
        # Try to parse quoted strings
        tokens = re.findall(r"""(['\"])(.*?)(?<!\\)\1""", inner)
        if tokens:
            return [t[1] for t in tokens]
        # Fallback to simple splitting
        parts = inner.split(',')
        result = []
        for part in parts:
            cleaned = part.strip().strip('"\'')
            if cleaned:
                result.append(cleaned)
        if result:
            return result
    
    return [s] if s else []


def speakers_to_roles(speakers: List[str]) -> Tuple[Optional[str], Optional[str]]:
    """
    Determine which unique speaker is the helper (counselor/therapist/etc.)
    and which is the other participant.
    Returns (helper_name, other_name) or (None, None) if ambiguous.
    """
    unique = []
    for sp in speakers:
        if sp not in unique:
            unique.append(sp)

    helper = None
    other = None
    for sp in unique:
        if is_helper_role(sp):
            helper = sp
            break
    if helper is None:
        return None, None

    for sp in unique:
        if sp != helper:
            other = sp
            break

    if other is None:
        return None, None

    return helper, other


def build_messages(dialogue: List[str], speakers: List[str]) -> Optional[List[Dict[str, str]]]:
    """
    Build OpenAI chat messages with roles:
      - helper (counselor/therapist) -> role "user"
      - other participant -> role "assistant"
    Uses only the dialogue text and speaker order.

    Returns None if invalid/misaligned or if no helper role is present.
    """
    if not dialogue or not speakers or len(dialogue) != len(speakers):
        print("length mismatch")
        return None

    helper, other = speakers_to_roles(speakers)
    print("helper, other:", helper, other)
    if helper is None or other is None:
        return None

    messages: List[Dict[str, str]] = []
    for sp, text in zip(speakers, dialogue):
        if not isinstance(text, str):
            continue
        role = "user" if sp == helper else "assistant"
        # Normalize whitespace a bit
        content = re.sub(r"\s+\n", "\n", text.strip())
        if content:
            messages.append({"role": role, "content": content})

    # Ensure the conversation starts with a 'user' message (recommended for stability)
    if messages and messages[0]["role"] != "user":
        messages = [{"role": "user", "content": ""}] + messages

    # Collapse any consecutive messages of the same role (defensive)
    collapsed: List[Dict[str, str]] = []
    for m in messages:
        if collapsed and collapsed[-1]["role"] == m["role"]:
            collapsed[-1]["content"] = (collapsed[-1]["content"] + "\n" + m["content"]).strip()
        else:
            collapsed.append(m)

    roles_present = {m["role"] for m in collapsed}
    if not {"user", "assistant"}.issubset(roles_present):
        return None

    return collapsed


def dataframe_to_examples(df: pd.DataFrame, dialogue_col: str, speakers_col: str) -> List[Dict]:
    """
    Convert a DataFrame into a list of {"messages": [...]} examples.
    Keeps only rows that contain a counselor/therapist in `speakers`.
    """
    results = []
    print("total rows:", len(df))
    for _, row in df.iterrows():
        dialogue_list = parse_dialogue(row.get(dialogue_col))
        speakers_list = parse_speakers(row.get(speakers_col))

        messages = build_messages(dialogue_list, speakers_list)
        if messages:
            print("message")
            results.append({"messages": messages})

    return results

```
